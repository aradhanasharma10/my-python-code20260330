```python
units = int(input("Enter units: "))

bill = 0

if units > 100:
  bill = bill + (min(units,200)-100) * 5

if units > 200:
  bill = bill + (units - 200) * 10


print("Total bill =", bill)
```

# Q4. Ask user for their salary and years of service and print the net bonus amount?


```python
salary = float(input("Enter salary: "))
years = int(input("Enter years of service: "))

if years > 10:
    bonus = salary * 10 / 100
elif years >= 6:
    bonus = salary * 8 / 100
else:
    bonus = salary * 5 /100

print("Bonus =", bonus)
```

# Q5. Accept three numbers from the user and display the second largest number?


```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

numbers = [a, b, c]
numbers.sort()
print("Second largest number is:", numbers[1])
```

# Q6. Marked Price Discount

```python
price = float(input("Enter Marked Price: "))

if price > 10000:
    discount = price * 20 / 100
elif price > 7000 and price <= 10000:
    discount = price * 15 / 100
else:
    discount = price * 10 / 100

net_amount = price - discount

print("Discount =", discount)
print("Net amount to pay =", net_amount)
```

# Q7. stream allotted according to the marks


```python
eng = int(input("Enter English marks: "))
math = int(input("Enter Math marks: "))
sci = int(input("Enter Science marks: "))
sst = int(input("Enter Social Studies marks: "))

if eng > 80 and math > 80 and sci > 80 and sst > 80:
    print("Science Stream")

elif eng > 80 and math > 50 and sci > 50:
    print("Commerce Stream")

elif eng > 80 and sst > 80:
    print("Humanities")

else:
    print("No stream allotted")
```

# Q9. last digit of a number (entered by user) is divisible by 3 or not?



```python
num = int(input("Enter a number: "))

last_digit = num % 10

if last_digit % 3 == 0:
    print("Last digit is divisible by 3")
else:
    print("Last digit is not divisible by 3")
```
